#### BANK CUSTOMER DATASET REGRESSION USING LINEAR,LASSO,RIDGE

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("bank_customers.csv")
data.head()
data.shape
```

```python
data
```

```python
x=data.drop("Savings",axis=1)
y=data["Savings"]
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test, y_train , y_test = train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
p = LinearRegression()
p.fit(x_train,y_train)
```

```python
yp=p.predict(x_test)
yp
```

```python
p.score(x_test,y_test)
```

```python
p.score(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rmse = np.sqrt(mean_squared_error(y_test,yp))
rmse
```

```python
from sklearn.linear_model import Lasso
```

```python
x1 = data.drop("Savings",axis=1)
y1=data["Savings"]
```

```python
k=Lasso()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x1, y1, test_size = 1/3, random_state = 0)
```

```python
k.fit(x_train,y_train)
```

```python
k.score(x_train,y_train)
```

```python
k.score(x_test,y_test)
```
